```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Plotting the periodicty for the anchor frame 0002 in attack dataset
# Read the CSV file without column names
df = pd.read_csv('D:\IIT-D\Sem-4\JCD893-M.Tech Major Project Part-2\Anchor_Frame\Dataset\Original/attack_dataset.csv', header=None)

# Filter rows where the second column value is '0002'
filtered_df = df[df[1] == '0002']

# Extract the first column (timestamps) and convert to a list
timestamps = filtered_df[0].tolist()

# Compute the timestamp differences
timestamp_diffs = [j - i for i, j in zip(timestamps[:-1], timestamps[1:])]

```

```python
# Plotting the timestamp differences as a bar graph
plt.figure(figsize=(25, 15))
plt.bar(range(len(timestamp_diffs)), timestamp_diffs)
plt.xlabel('Index')
plt.ylabel('Timestamp Difference')
plt.title('Periodicity for CAN ID 0002')
max_diff = max(timestamp_diffs)
plt.yticks(np.arange(0, max_diff, 0.05), fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.9)
plt.show()
```

```python

plt.figure(figsize=(20, 10))
plt.bar(np.arange(len(timestamp_diffs)), timestamp_diffs, color=(0.2, 0.4, 0.6, 0.6))
plt.xlabel('Index')
plt.ylabel('Timestamp Difference')
plt.title('Periodicity for CAN ID 0002')
timestamp_diffs = np.array(timestamp_diffs)
plt.ylim(0,np.average(timestamp_diffs)+np.std(timestamp_diffs))
plt.grid(axis='y', linestyle='--', alpha=0.9)
plt.show()
```
