```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tensorflow.keras.models import Sequential, load_model
```

```python
## Car Hacking - 0.095
##MCAN - 0.018
```

```python
mlp_dos = load_model('../data/M-CAN Intrusion Dataset/DoS-to-Adversarial Transfer Learning Evaluation/mlp.h5')
mlp_robust = load_model('../data/M-CAN Intrusion Dataset/exp_factor_0.018/Adversarial Training Evaluation/mlp.h5')
```

```python
dos_weights = mlp_dos.layers[2].get_weights()[0]
robust_weights = mlp_robust.layers[2].get_weights()[0]
```

```python
print(dos_weights.shape)
print(robust_weights.shape)
```

```python
dos_weights = mlp_dos.layers[0].get_weights()[0]
robust_weights = mlp_robust.layers[0].get_weights()[0]
```

```python
dos_weights1 = dos_weights[0].reshape(-1,1)
robust_weights1 = robust_weights[0].reshape(-1,1)

dos_weights2 = dos_weights[1].reshape(-1,1)
robust_weights2 = robust_weights[1].reshape(-1,1)

dos_weights3 = dos_weights[2].reshape(-1,1)
robust_weights3 = robust_weights[2].reshape(-1,1)

dos_weights4 = dos_weights[3].reshape(-1,1)
robust_weights4 = robust_weights[3].reshape(-1,1)
```

```python
print(dos_weights2.shape)
print(robust_weights2.shape)
```

```python
def weights_dataframe(dos_weights, robust_weights):

    weight_dict = {}

    dim = dos_weights.shape[0]

    weight_dict = {'DoS_Weights':[], 'Robust_Weights':[], 'Diff':[]}

    for i in range(dim):
        weight_dict['DoS_Weights'].append(dos_weights[i][0])
        weight_dict['Robust_Weights'].append(robust_weights[i][0])
        weight_dict['Diff'].append(dos_weights[i][0] - robust_weights[i][0])


    print("Original Model Weight Value Statistics:")
    print("Min: ",np.min(weight_dict['DoS_Weights']))
    print("Median: ",np.median(weight_dict['DoS_Weights']))
    print("Max: ",np.max(weight_dict['DoS_Weights']))

    print()
    print("-------x-------")
    print()

    print("Robust Model Weight Value Statistics:")
    print("Min: ",np.min(weight_dict['Robust_Weights']))
    print("Median: ",np.median(weight_dict['Robust_Weights']))
    print("Max: ",np.max(weight_dict['Robust_Weights']))

    return pd.DataFrame(weight_dict)
    
```

```python
mlp_dataframe = weights_dataframe(dos_weights, robust_weights)

mlp_dataframe[['DoS_Weights', 'Robust_Weights']].plot(kind='bar', figsize=(30,20))
plt.title("M-CAN Dataset: DoS vs Robust Weights (MLP : Layer 3)") 
plt.ylim(-5, 5)
```

```python
mlp1_dataframe = weights_dataframe(dos_weights1, robust_weights1)
```

```python
## Write code to plot a bar graph comparing the weight values of the DoS and Robust models

mlp1_dataframe[['DoS_Weights', 'Robust_Weights']].plot(kind='bar', figsize=(30,20))
plt.title("M-CAN Dataset: DoS vs Robust Weights (MLP : Layer 1, 1st Feature)") 
plt.ylim(-2.5, 2.5)
```

```python
mlp2_dataframe = weights_dataframe(dos_weights2, robust_weights2)

## Write code to plot a bar graph comparing the weight values of the DoS and Robust models

mlp2_dataframe[['DoS_Weights', 'Robust_Weights']].plot(kind='bar', figsize=(30,20))
plt.title("M-CAN Dataset: DoS vs Robust Weights (MLP : Layer 1, 2nd Feature)")
plt.ylim(-2.5, 2.5) 
```
